```python
res = res.drop(columns=[
 "LVREQSTYEAR",
 "LVREQSTID",
 "LVREQSTDT",
 "LVFRDT",
 "LVTODT"
])
```

```python
res.shape
```

```python
#for categorical values since there is no constant values ..so this approach followed

res["LVREASON"] = res["LVREASON"].fillna("UNKNOWN")
res["ADNLINFO"] = res["ADNLINFO"].fillna("NONE")
res["HALFDAY"] = res["HALFDAY"].fillna("NO")
```

```python
#Numeric values missing

num_cols = ["ELBALANCE", "HPLBALANCE", "leave_days"]
for col in num_cols:
 res[col] = res[col].fillna(res[col].median())
```

```python
res = res[res["leave_days"] > 0]
```

```python
res.to_csv('CLEANED_DATASET.csv',index=False)
```
